```python
import pandas as pd
import numpy as np
```

```python
raw = pd.read_csv('./input.csv')
```

```python
main = raw.copy()
```

```python
## part 1
```

```python
main['prev'] = main.shift()
```

```python
main['change'] = np.where(main.value > main.prev, 1, 0)
```

```python
main.change.sum()
```

```python
## part 2
```

```python
main.iloc[0].value
```

```python
li = []
for i in range(len(main)-2):
    a = i
    b = i+1
    c = i+2
    temp = main.iloc[a].value + main.iloc[b].value + main.iloc[c].value
    li.append(temp)
```

```python
li[1997]
```

```python
main.drop(main.tail(2).index, inplace=True)
```

```python
main['cumm'] = li
```

```python
main['prev2'] = main.cumm.shift()
main['change2'] = np.where(main.cumm > main.prev2, 1, 0)
```
